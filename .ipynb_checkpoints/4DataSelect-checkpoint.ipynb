# 数据选择

## 列选择

### 选择某一列、某几列

* 普通索引
在df后面的方括号指明要选择的列名，如果一列，只需要传入一个列名  
同时选择多列，则传入多个列名  
> 这里需要注意传入的参数问题  
> `df['性别']`  &  `df[['姓名','性别']]` & `df.loc[:,['姓名','性别']]`

* 位置索引
在df后面的iloc方法指明指定具体列的位置，即第几列，对数据进行选取，通过传入位置来获取数据  
> `df.iloc[:,[0,2]]`

```python
df.loc[:,['性别','姓名']].head()
```

```python
import pandas as pd

df = pd.read_excel('./datapool/岭南师范学院第二针.xls')
df[['姓名','性别']].head()
```

### 选择连续某几列
选择连续的某几列：用到iloc方法，只要传入这些连续列的位置区间即可

> `df.iloc[:,0:5]`  
> 这里需要注意，列索引包含0列，不包含第五列，行从第一行开始算起，不包括索引列

```python
df.iloc[:,[4,7]]
```

```python
df.iloc[:,0:6].head()
```

##  行选择

### 选择某一行/某几行

* 普通索引  
即传入具体**索引的名称**，用到loc方法  
选择一行  `df.loc[25]`  
选择多行  `df.loc[[1,2,3]]`  
* 位置索引  
即传入**具体的行数**，用到iloc方法  
选择一行，`df.iloc[0]`  
选择多行，`df.iloc[[0,1]]`  

```python
df.loc[25]
```

```python
df.loc[[0,25,50]]
```

```python
df.iloc[[0,25,50]]
```

### 连续选择某几行

使用iloc的方法，把连续行的位置用一个区间表示    
`df.iloc[0:5]`

```python
df.iloc[0:5]
```

###  选择满足条件的行
直接在表名后面指明哪列要满足什么条件  
`df[df['第一剂时间'] == '2021-03-29']`  
也可以指定多个条件进行过滤  
每个条件用一个圆括号括起来，然后用操作符进行连接，完成想要达到的操作  

```python
df[(df["第一剂时间"] == '2021-03-29') | (df['第一剂时间'] == '2021-04-01')].head()
```

```python
ii = '1996-12-31'
df[(df['出生日期'] < ii) & (df['性别'] == '男')]
```

## 行列同时选择

###  普通索引+普通索引选择指定的行和列

使用loc方法，同时传入行和列的**具体的索引名称**进行数据选择  

```python
df.loc[[1,25,50],['姓名','出生日期','第一剂时间']]
```

###  位置索引+位置索引选择指定的行和列

使用iloc方法，**传入行、列索引的位置**来获取数据

```python
df.iloc[[1,25,50],[4,6,9]]
```

### 布尔索引+普通索引选择指定的行和列
先对表格进行布尔索引选择行，然后通过普通索引选择列  
其实先对表格进行布尔索引会得到一个DataFrame类型的表格，其值全为满足条件的True     
然后在对这个表格进行普通索引，所谓普通索引就是指定表格的列命，在前面的基础上继续得到一个新的表格  
> 这里需要注意的地方跟之前的一样，得到的一个DataFramel类型的表格，想得到里面的多个列  
> 要注意需要两个括号`df[['姓名'],['出生日期']]`

```python
h = '1997-12-31'
df[(df['出生日期'] < h)][['姓名', '出生日期']].head()
```

###  切片索引+切片索引选择指定的行和列
切片索引+切片索引是通过同时传入行、列索引的位置区间进行数据选择  

```python
df.iloc[[1,25,50],[1,4,9]]
```
