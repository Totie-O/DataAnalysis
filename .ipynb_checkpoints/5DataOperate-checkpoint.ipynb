## 数值操作

### 数值替换
* 利用replace(A,B)方法，将A替换成B  
`df['学生姓名'].replace('王南强','neoqiang',inplace = True)`   
* 多对一替换  
`df.['学生姓名'].replace(['王南强','黄佩诗','黄轩娣'],'neoqiang')`
* 多对多替换
`df.replace({'王南强':'neoqiang','黄佩诗':'neohuang'})`

````python
import pandas as pd
import numpy as np

df = pd.read_excel(r'datapool/受助名单.xls',header = 3)
````

````python
df.replace({'王南强':'neoqiang','黄佩诗':'neohuang'},inplace = True)
df
````

### 1.3 数值排序
数值排序是按具体数值的大小进行排序，有升序和降序，升序就是数值由小到大排列，降序是数值由大到小排列   
`df.sort_values(by=['coll'],ascending=False,na_position='last')`  
* 上面代码表示df表按照coll列进行排序，ascending=False表示按照coll列进行降序排序  
   ascending参数默认值为True,表示升序排列  
*  na_pasition对缺失值进行设置，默认参数值为'last',可以不写，表示将缺失值显示在最后  
   设置na_position参数为'first',将缺失值显示在最前面   
   
`df.sort_values(by=['coll1','coll2'],ascending=[True,False],na_position='last')`  
*  同样使用sort_values可以进行多列排序，只需要在sort_values后的by以列表的形式指明要排序的多列列名及每列的排序方式即可。  

````python
import pandas as pd

df1 = pd.read_excel(r'datapool/岭南师范学院第二针.xls')
df1.sort_values(by = ['出生日期']).head()
````

### 1.4 数值排名
`rank(ascending=True,method='average')`  
average:当排名有重复值时，返回重复值的平均排名  
first:按值在所有的待排列数据中出现的先后顺序排名，**重复值第一次出现的为一，重复值第二次出现的为二**  
min:返回重复值的最佳排名  
max:与min相反，取重复值对应的最大排名  

````python
import pandas as pd

df2 = pd.read_excel(r"datapool/search-grade.xlsx")
df2['笔试成绩总分'].rank(method='min')
````

### 1.5 数值删除之删除列
Python用到`drop`方法，drop方法后面的括号中指明要删除的列名或者列的位置，即第几列  
* 在drop方法后的括号中直接传入待删除列的列名，需要加一个参数axis，并让其参数值等于1，表示删除列  
* 还可以直接传入待删除列的位置，但也需要用axis参数  
* 也可以将列名以列表的形式传给columns参数，这个时候就不需要axis参数了
```python
df1.drop(['第二剂时间','第二剂厂家'],axis=1)
df1.drop(df1.columns[[11,12]],axis=1)
df1.drop(columns = ['第二剂时间','第二剂厂家'])
```
> 这里需要注意，`df1.columns[[11,12]]`返回`Index(['第二剂时间', '第二剂厂家'], dtype='object')`

````python
import pandas as pd

df1 = pd.read_excel(r'datapool/岭南师范学院第二针.xls')
df1.drop(['第二剂时间','第二剂厂家'],axis = 1)
````

````python
df1.drop(df1.columns[[11,12]],axis=1)
````

````python
print(df1.columns)
print(df1.index)
````

### 1.6 数值删除之删除行
Python中删除某些行用到的方法依然是drop(),与删除列类似的是，**删除行也要指明行相关的信息**  
* 在drop方法后的括号中直接传入待删除行的行名，并让axis参数值等于0，表示删除行  
* 除了传入行索引名称，还可以在drop方法后的括号中直接传入待删除行的行号，也需要用axis参数并让其参数值等于0  
* 可以将带删除的行名传给index参数，这个时候就不需要axis参数了  
```python
df3=df.drop([0],axis=0)
df3=df.drop(df.index[[0]],axis=0 & df3=df.drop(df.index[0],axis=0)
df3=df.drop(index=[0])
```

````python
import pandas as pd
import numpy as np
df = pd.read_excel(r'datapool/受助名单.xls',header=3)
df3 = df.drop(df.index[[0,19,20,21,22,23]],axis=0)
df3
````

### 1.10 数值查找
数值查找就是查看数据表中的数据是否包含某个值或者某些值  
在Python中查看数据表中是否包含某个值用到的是isin()方法，而且可以同时查找多个值  
只需要在isin方法后的括号指明即可  
可以将某列数据取出来，然后在这一列上调用isin()方法，看这一列是否包含某个/些值，如果包含则返回True，否则返回False

```python
df3['家庭经济困难等级'].isin(['一般困难'])
df3['家庭经济困难等级'].isin(['一般困难','特殊困难'])
df3['家庭经济困难等级'] == '一般困难'
df3['家庭经济困难等级'] == ('一般困难' or '特殊困难')
```

````python
df3['家庭经济困难等级'].isin(['一般困难','特殊困难'])
````

````python
df3['家庭经济困难等级'] == ('一般困难' or '特殊困难')
````

### 1.11 区间切分
区间切分就是将一系列值分成若干份  
Python中对区间切分利用的是cut()方法，cut方法有一个参数bin用来指名切分区间
类似的还有qcut()方法，不需要事先指定切分区间，只需要指名切分个数

````python
import pandas as pd

df2 = pd.read_excel(r"datapool/search-grade.xlsx")
pd.qcut(df2['笔试成绩总分'],10)
````

### 1.12 插入新的行或列   
在python中插入一个新的列用到的方法是insert(),在insert方法后的括号中指明要插入的位置、插入后新列的列名，以及要插入的数据  
```python
df6 = df2['笔试成绩总分'].rank(method='min')
df2.insert(2,'笔试总成绩排名',df6.astype('int')
df2['笔试成绩排名'] = df6.astype('int')
```
还可以直接以索引的方式进行行列的插入，直接让新的一列等于某列值即可

````python
import pandas as pd

df2 = pd.read_excel(r"datapool/search-grade.xlsx")
df6 = df2['笔试成绩总分'].rank(method='min')

df2['笔试成绩排名'] = df6.astype('int')
df2.head(2)
````

### 1.14 索引重塑
所谓索引重塑就是将原来的索引进行重新构造  
树形结构其实就是维持表格型行索引不变的前提下，把列索引也变成行索引，  
其实就是给表格型数据建立层次化索引  
数据从表格型数据转换到树形数据的过程叫重塑  
与stack()方法相对应的方法是unstack()方法，  
unstack()方法是将树形结构转为表格型数据  

````python
df3.stack().unstack()
````

### 1.15 apply()与applymap()函数
函数apply()和applymap()都需要与匿名函数lambda结合使用  
apply()函数主要用于对DataFrame中某一column或row中的元素执行相同的函数操作  
applymap()函数用于对DataFrame中的每一个元素执行相同的函数操作  


````python
df3['序号'] = df3['序号'].apply(lambda x:int(x)-1)
````
