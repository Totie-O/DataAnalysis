### 数据分组

Python中对数据分组利用的是groupby()方法,这个有点类似sql中的groupby
```python
df.groupby('专业')
df.groupby(['专业','班级'])
```
都可以对分组后的df进行聚合运算  

````python
import pandas as pd

df = pd.read_excel(r'datapool/search-grade.xlsx')
````

````python
df.groupby(['班级']).count()
````

````python
df.groupby('专业')['序号'].count()
````

````python
df.groupby('班级')['序号'].count()['2019金融工程5班']
````

### 分组键是Series

把DataFrame的其中一列取出来就是一个Series，比如下面的  
`df['班级']`  
分组键是列名与分组键是Series的唯一区别就是，给groupby()方法传入了什么  
其他都一样。可以按照一个或多个Series进行分组，分组以后的汇总计算也是完全一样的   
也支持对分组后的某些列进行汇总计算  
`df.groupby([df['专业'],df['班级']])`

````python
df.groupby(df['班级']).count()
````

````python
df.groupby([df['专业'],df['班级']]).count()
````

### 神奇的aggregate方法
前面用到的聚合函数都是直接在DadaFrameGroupBy上调用，这样分组以后所有列做的都是同一种运算，  
且一次只能使用一种汇总方式   
aggregate的第一个神奇之处在于，一次可以使用多种汇总方式，  
aggregate第二个神奇之处在于，可以针对不同的列做不同的汇总运算  

````python
df.groupby('专业')['笔试成绩总分'].aggregate(['count','sum'])
````

````python
df1 = df.groupby(['专业','班级']).aggregate({'序号':'count','听力分数':'mean','阅读分数':'mean','作文分数':'mean'})
df1.reset_index(inplace=True)
df1
````
