## 数值操作

### 数值替换
* 利用replace(A,B)方法，将A替换成B  
`df.loc[:,0].replace(np.NaN, '5月10日晚上7:30-9:30',inplace=True)`   
* 多对一替换  
`df.['学生姓名'].replace(['王南强','黄佩诗','黄轩娣'],'neoqiang')`
* 多对多替换
`df.replace({'18酒店':'18酒店管理','18物流':'18物流管理','18工商':'18工商管理'})`

````python
import pandas as pd
import numpy as np

df = pd.read_excel('./datapool/期末考试安排（商学院）.xlsx',header=None)

df.replace({'18酒店':'18酒店管理','18物流':'18物流管理','18工商':'18工商管理'},inplace=True)
#df
df.loc[:,0].replace(np.NaN, '5月10日晚上7:30-9:30',inplace=True)
#df
df.loc[:,3].replace([34,26,35],99,inplace=True)
df
````

### 数值排序
数值排序是按具体数值的大小进行排序，有升序和降序，升序就是数值由小到大排列，降序是数值由大到小排列   
```python
df.sort_values(by=['coll'],ascending=False,na_position='last')
```  
* 上面代码表示df表按照coll列进行排序，ascending=False表示按照coll列进行降序排序  
   ascending参数默认值为True,表示升序排列  
*  na_pasition对缺失值进行设置，默认参数值为'last',可以不写，表示将缺失值显示在最后  
   设置na_position参数为'first',将缺失值显示在最前面   
   
```python
df.sort_values(by=['coll1','coll2'],ascending=[True,False],na_position='last')
```  
*  同样使用sort_values可以进行多列排序，只需要在sort_values后的by以列表的形式指明要排序的多列列名及每列的排序方式即可。

````python
import pandas as pd

df2 = pd.read_excel(r'./datapool/岭南师范学院第二针.xls')
df2.sort_values(by = ['出生日期']).head()
````

### 数值排名

`rank(ascending=True,method='average')`  
1. average:当排名有重复值时，返回重复值的平均排名  
2. first:按值在所有的待排列数据中出现的先后顺序排名，**重复值第一次出现的为一，重复值第二次出现的为二**  
3. min:返回重复值的最佳排名  
4. max:与min相反，取重复值对应的最大排名  

````python
import pandas as pd

df3 = pd.read_excel("./datapool/search-grade.xlsx")

df3['总成绩排名'] = df3['笔试成绩总分'].rank(method='min')
df3.head(2)
````

### 数值删除之删除列
Python用到`drop`方法，drop方法后面的括号中指明要删除的列名或者列的位置，即第几列  
* 在drop方法后的括号中直接传入**待删除列的列名**，需要加一个参数axis，并让其参数值等于1，表示删除列  
* 还可以直接传入待**删除列的位置**，但也需要用axis参数  
* 也可以将**列名以列表的形式**传给columns参数，这个时候就不需要axis参数了
```python
df3.drop(['第二剂时间','第二剂厂家'],axis=1)
df3.drop(df1.columns[[11,12]],axis=1)
df3.drop(columns = ['第二剂时间','第二剂厂家'])
```
> 这里需要注意，`df1.columns[[11,12]]`返回`Index(['第二剂时间', '第二剂厂家'], dtype='object')`

````python
import pandas as pd

df3 = pd.read_excel('./datapool/岭南师范学院第二针.xls')
df3.drop(['第二剂时间','第二剂厂家'],axis = 1).head()
````

### 数值删除之删除行
Python中删除某些行用到的方法依然是drop(),与删除列类似的是，**删除行也要指明行相关的信息**  
* 在drop方法后的括号中直接**传入待删除行的行名**，并让axis参数值等于0，表示删除行  
* 除了传入行索引名称，还可以在drop方法后的括号中直接传入**待删除行的行号**，也需要用axis参数并让其参数值等于0  
* 可以将带**删除的行名传给index参数**，这个时候就不需要axis参数了  
```python
df3.drop([0,1,2],axis=0)
df3.drop(df.index[[0,1,2,3]],axis=0) 
df3.drop(index=[0])
```

### 数值查找

数值查找就是查看数据表中的数据是否包含某个值或者某些值  
在Python中查看数据表中是否包含某个值用到的是isin()方法，而且可以同时查找多个值  
只需要在isin方法后的括号指明即可  
可以将某列数据取出来，然后在这一列上调用isin()方法，看这一列是否包含某个/些值，如果包含则返回True，否则返回False

```python
df3['第一剂时间'].isin(['2021-03-29'])
df3['第一剂时间'].isin(['2021-03-29','2021-04-01'])
df3['第一剂时间'] == '2021-03-29'
df3['第一剂时间'] == ('2021-03-29' or '2021-04-01')
```

````python
df3['第一剂时间'].isin(['2021-03-29','2021-04-01'])
````

### 插入新的行或列  

 在python中插入一个新的列用到的方法是insert(),在insert方法后的括号中指明要插入的位置、插入后新列的列名，以及要插入的数据

### 索引重塑
所谓索引重塑就是将原来的索引进行重新构造  
树形结构其实就是维持表格型行索引不变的前提下，把列索引也变成行索引，  
其实就是给表格型数据建立层次化索引  
数据从表格型数据转换到树形数据的过程叫重塑  
与stack()方法相对应的方法是unstack()方法，  
unstack()方法是将树形结构转为表格型数据  

###  apply()与applymap()函数

函数apply()和applymap()都需要与匿名函数lambda结合使用  
apply()函数主要用于对DataFrame中某一column或row中的元素执行相同的函数操作  
applymap()函数用于对DataFrame中的每一个元素执行相同的函数操作  

````python
df6.reset_index(inplace=True)
df6['人数'] = df6['人数'].apply(lambda x:int(x)+100)
````
