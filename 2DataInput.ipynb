## 指定传入的表格

当excel文件有多个表格时，可以用sheet_name参数指定相应的表格，仅需要将表格的名字传入给sheet_name，还可以传入数字0、1、2指定相应表格，表格按顺序排列   

`df = pd.read_excel(r'./datapool/期末考试安排（商学院）.xlsx', sheet_name = '专业课')`  
`df = pd.read_excel(r'./datapool/期末考试安排（商学院）.xlsx', sheet_name = 2)`

## 指定行索引
index_col默认从0开始，可以通过指定index_col的值来指定列做为行索引

`df = pd.read_excel(r'./datapool/期末考试安排（商学院）.xlsx', index_col = 0)`

```python
df = pd.read_excel('./datapool/期末考试安排（商学院）.xlsx', index_col = 0)
df
```

## 指定列索引  
将本地文件导入DataFrame中，**默认使用源数据表的第一行做为列索引**  
因为header参数值默认为0，即用第一行做为列索引  
可以通过header参数来设置列索引  

`df = pd.read_excel(r'datapool/期末考试安排（商学院）.xlsx', header = 1)`

```python
df = pd.read_excel(r'datapool/期末考试安排（商学院）.xlsx', header = 1)
df
```

还可以指定header为**None**,当header为None时，去掉源数据表格的第一行所有数据，但是保留位置，值从0开始  

```python
df = pd.read_excel(r'datapool/期末考试安排（商学院）.xlsx', header = None)
df
```

## 指定导入的列
通过指定usecols的参数来指定要导入的列 
usecols指定的参数是一个列表，不能只单纯指定一个列，需要指定两个列以上，用一个列表包裹    

`df = pd.read_excel(r'datapool/期末考试安排（商学院）.xlsx', usecols=[0,1,2,3,4,5])`

```python
df = pd.read_excel(r'datapool/期末考试安排（商学院）.xlsx', usecols=[0,1,2,3,4])
df
```
